## Actual Analysis Begins Here

```python
# grab critical columns
attr_df = df[['iid', 
              'attr1_1', 'attr1_2', 
              'sinc1_1', 'sinc1_2', 
              'intel1_1', 'intel1_2', 
              'fun1_1', 'fun1_2', 
              'amb1_1', 'amb1_2',
              'shar1_1', 'shar1_2']].groupby('iid').mean()
```

```python
attr_df.shape
```

```python
clean_attr = attr_df.dropna()
clean_attr.shape
clean_attr.head()
```

### Correlation Analysis

```python
clean_attr.corr()
```

```python
attr_before = clean_attr[['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']]
attr_after = clean_attr[['attr1_2', 'sinc1_2', 'intel1_2', 'fun1_2', 'amb1_2', 'shar1_2']]
attr_after.head()
```

```python
attr_after.columns = ['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']
attr_after.head()
```

```python
isolated_corr = attr_before.corrwith(attr_after)
isolated_corr
```

```python
isolated_corr.to_csv('correlations.csv')
```

### Vector Analysis

```python
vector_attr = pd.DataFrame(
    { 
        't1': clean_attr[['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']].values.tolist(), 
        't2': clean_attr[['attr1_2', 'sinc1_2', 'intel1_2', 'fun1_2', 'amb1_2', 'shar1_2']].values.tolist()
    
    })
vector_attr.head()
```

```python
def cosine_dist(row):
    return scipy.spatial.distance.cosine(row['t1'], row['t2'])
```

```python
vector_attr['cosine_dist'] = vector_attr.apply(cosine_dist, axis=1)
vector_attr.head()
```

```python
vector_attr.head().values
```

```python
cosine_dists = vector_attr['cosine_dist']
cosine_dists.shape
cosine_dists.head()
cosine_dists.hist()
```

```python
cosine_dists.to_csv('cosine_dists.csv')
```

```python
results = scipy.stats.ttest_1samp(cosine_dists, 0)
```

```python
results.statistic
```

```python
results.pvalue
```

```python
results.pvalue / 2
```
